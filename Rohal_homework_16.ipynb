## **1. Створення масиву та операції**
Створіть два двовимірних масиви розміром 5x5 кожен, заповніть їх випадковими числами. Потім виконайте наступні операції та виведіть результат:
* Сума елементів кожного масиву.
* Різниця між елементами двох масивів.
* Перемноження елементів двох масивів.
* Підняття кожного елемента першого масиву до ступеня відповідного елемента другого масиву.

```python
random_array_1 = np.random.randint(50, size=(5, 5))
random_array_1
```

```python
random_array_2 = np.random.randint(50, size=(5, 5))
random_array_2
```

### Сума елементів кожного масиву.

```python
array_sum = random_array_1 + random_array_2
array_sum
```

### Різниця між елементами двох масивів.

```python
array_sub = random_array_1 - random_array_2
array_sub
```

### Перемноження елементів двох масивів.

```python
array_mult = random_array_1 * random_array_2
array_mult
```

### Підняття кожного елемента першого масиву до ступеня відповідного елемента другого масиву.

```python
array_exp = np.power(random_array_1, random_array_2)
array_exp
```

## **2. Індексація та вибірка даних** 
Створіть двовимірний масив 8x8, який відображає шахову дошку (1 - біла клітина, 0 - чорна клітина). Використовуйте індексацію для виведення на екран:
* Рядок, представляючи 3-й рядок з дошки.
* Стовпець, представляючи 5-й стовпець з дошки.
* Підмасив, представляючи частину дошки розміром 3x3 в лівому верхньому куті.

```python
chess_board = np.zeros((8, 8), dtype=int)
chess_board[::2, 1::2] = 1
chess_board[1::2, ::2] = 1
chess_board
```

### Рядок, представляючи 3-й рядок з дошки.

```python
row_3 = chess_board[2, :]
row_3
```

### Стовпець, представляючи 5-й стовпець з дошки.

```python
col_5 = chess_board[:, 4]
col_5
```

### Підмасив, представляючи частину дошки розміром 3x3 в лівому верхньому куті.

```python
small_chessboard = chess_board[:3, :3]
small_chessboard
```

### Знайдіть середнє значення, мінімум та максимум для кожного каналу зображення (R, G, B).

```python
rgb_avg = np.mean(img_dog, axis=0)
np.mean(rgb_avg, axis=0)
```

```python
rgb_max = np.max(img_dog, axis=0)
np.max(rgb_max, axis=0)
```

```python
rgb_min = np.min(img_dog, axis=0)
np.min(rgb_min, axis=0)
```

### Підрахуйте загальну суму інтенсивності пікселів та виведіть її.

### Виконайте нормалізацію зображення, розділивши значення кожного пікселя на максимальне значення.

```python
bw_img_dog = cv2.cvtColor(img_dog, cv2.COLOR_BGR2GRAY)
```

```python
normalized_bw_img = cv2.normalize(
    bw_img_dog, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
```

```python
normalized_color_img = cv2.cvtColor(
    normalized_bw_img, cv2.COLOR_GRAY2BGR)
```

```python
plt.imshow(normalized_color_img)
plt.show()
```
